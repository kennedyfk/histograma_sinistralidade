```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Carregar o CSV (padro BR: ISO-8859-1, sep=';', decimal=',')
df = pd.read_csv('_indicadores_consolidado.csv', encoding='ISO-8859-1', sep=';', decimal=',')

# Seleciona apenas as colunas necessrias
dados_intervalo = df[['CD_OPERADORA', '03_is', 'INICIO_TRIMESTRE', 'FIM_TRIMESTRE']].dropna()

# Converter para datetime (se ainda no for)
dados_intervalo['INICIO_TRIMESTRE'] = pd.to_datetime(dados_intervalo['INICIO_TRIMESTRE'], dayfirst=True, errors='coerce')
dados_intervalo['FIM_TRIMESTRE'] = pd.to_datetime(dados_intervalo['FIM_TRIMESTRE'], dayfirst=True, errors='coerce')


# Filtra os valores vlidos entre 0 e 2 na coluna '03_is'
dados_intervalo = dados_intervalo[(dados_intervalo['03_is'] >= 0) & (dados_intervalo['03_is'] <= 2)]

# Agora voc pode usar normalmente:
mu = dados_intervalo['03_is'].mean()
sigma = dados_intervalo['03_is'].std()


# 3. Parmetros de cor
bg_color    = '#04212b'   # fundo escuro
bar_color   = '#ff7f0e'   # laranja vibrante
edge_color  = '#ffffff'   # borda branca
line_color  = '#17becf'   # ciano para a curva normal
text_color  = '#ffffff'   # rtulos em branco
grid_color  = '#ffffff'   # grade branca, mas baixa opacidade

# 4. Criar figura e eixos
fig, ax = plt.subplots(figsize=(10, 6), facecolor=bg_color)
ax.set_facecolor(bg_color)

# 5. Histograma da coluna '03_is'
ax.hist(
    dados_intervalo['03_is'],
    bins=75,
    range=(0, 2),
    density=True,
    color=bar_color,
    edgecolor=edge_color
)

# 6. Curva da Normal
x = np.linspace(0, 2, 200)
pdf = stats.norm.pdf(x, mu, sigma)
ax.plot(x, pdf, color=line_color, lw=2, label='Distribuio Normal')

# 7. Ajustes de rtulos, ttulo e ticks
ax.set_title(
    "Histograma Detalhado do 3. Sinistralidade da Resoluo Normativa N 518 (ANEXO IV)",
    pad=30,
    color=text_color,
    y=0.95
)
ax.set_xlabel("Sinistralidade", color=text_color)
ax.set_ylabel("Densidade", color=text_color)
ax.set_xticks(np.arange(0, 2.1, 0.10))
ax.tick_params(axis='x', colors=text_color)
ax.tick_params(axis='y', colors=text_color)

# 8. Grade mais discreta
ax.grid(
    True,
    which='both',
    linestyle='--',
    linewidth=0.5,
    color=grid_color,
    alpha=0.3
)

# 9. Spines em branco
for spine in ax.spines.values():
    spine.set_color(text_color)

# 10. Legenda com fundo transparente
leg = ax.legend(frameon=True, facecolor=bg_color, edgecolor=edge_color)
plt.setp(leg.get_texts(), color=text_color)

plt.show()

print("\n")
# Sinistralidade Mdia
mu = dados_intervalo['03_is'].mean()
print(f"Sinistralidade Mdia: {mu:.4f}".replace(",", "."))

# Desvio Padro
sigma = dados_intervalo['03_is'].std()
print(f"Desvio Padro: {sigma:.4f}".replace(",", "."))

print("\n")
# Perodo Histrico de Anlise
min_data = dados_intervalo['INICIO_TRIMESTRE'].min()
max_data = dados_intervalo['FIM_TRIMESTRE'].max()
print(f"Perodo Histrico de Anlise: {min_data.strftime('%m/%Y')} at {max_data.strftime('%m/%Y')}")

# Qtde. de Sinist. Analisadas
qtde_sinist = len(dados_intervalo)
print(f"Qtde. de Sinist. Analisadas: {qtde_sinist:,} Sinistralidades".replace(",", "."))

# Qtde. de Operadoras Analisadas
qtde_operadora = dados_intervalo['CD_OPERADORA'].nunique()
print(f"Qtde. de Operadoras Analisadas: {qtde_operadora}")
```
